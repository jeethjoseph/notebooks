# Cosine Similarity

```python
matrix = data.pivot_table(values='qty',index ='itemID', columns ='bill')
matrix.head()
```

```python
matrix.shape
```

```python
matrix_dummy = matrix.copy().fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(matrix_dummy, matrix_dummy)
```

```python
cosine_sim = pd.DataFrame(cosine_sim, index = matrix.index,columns = matrix.index)
```

```python
cosine_sim.head()
```

```python
matrix_dummy.loc[[88486, 62602]].transpose().to_csv("out.csv")
```

# Final Function

```python
# Final Function
# Takes title, userId and returns to 25 sorted movie names on est
def reco(idx):

    similar_itemids = cosine_sim.loc[idx].to_frame(name=None)
    topItemIds = similar_itemids.sort_values(by =[idx],ascending = False).head(26)
    topItemNames = topItemIds.merge(items_index, how ='left', on = 'itemID' )

    
    return topItemNames.head(10)

```
