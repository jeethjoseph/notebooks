```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/movies_metadata.csv')
df.columns
```

```python
df = df[['title', 'genres','release_date','runtime','vote_average','vote_count']]
```

```python
df.head()
```

```python
df['release_date']=pd.to_datetime(df['release_date'],errors='coerce')
df.head()
```

```python
df['year']=df['release_date'].apply(lambda x:str(x).split('-')
                                    [0]if x!=np.nan 
                                    else np.nan)
df.head()
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return 0
    
```

```python
df['year']=df['year'].apply(convert_int)
```

```python
df.head()
```

```python
df=df.drop('release_date',axis=1)
df.head()
```

```python
df.iloc[0]['genres']
```

```python
from ast import literal_eval
```

```python
a=literal_eval(df.iloc[0]['genres'])
print(a)
```

```python
df['genres'] = df['genres'].fillna('[]')
```

```python
df['genres'] = df['genres'].apply(literal_eval)
```

```python
df['genres'] = df['genres'].apply(lambda x:[i['name']for i in x] if isinstance(x,list) else[])
```

```python
df.head()
```

Exploding the genres to make multiple movie entries. Movie names no longer unique

```python
s = df.apply(lambda x: pd.Series(x['genres']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
s.name = 'genre'
```

```python
gen_df = df.drop('genres', axis =1).join(s)
```

```python
gen_df.head()
```

```python
def build_chart(gen_df, percentile=0.8):
    print("Input preferred genre")
    genre = input()
    
    print("Input shortest duration")
    low_time = int(input())
    
    print("Input longest duration")
    high_time = int(input())
    
    print("Input earliest year")
    low_year = int(input())
    
    print("Input latest year")
    high_year =int(input())
    
    movies = gen_df.copy()
    movies = movies[(movies['genre'] == genre) & (movies['runtime'] >= low_time) & (movies['runtime'] <= high_time) & (movies['year'] >= low_year)&(movies['year'] <= high_year)]
    
    c = movies['vote_average'].mean()
    m = movies['vote_count'].quantile(percentile)
    
    q_movies = movies.copy().loc[movies['vote_count'] >= m]
    
    q_movies['score'] = q_movies.apply(lambda x:((x['vote_count']*x['vote_average']+m*c)/(x['vote_count']+m)), axis = 1)
    
    q_movies = q_movies.sort_values('score',ascending=False)
    
    return q_movies
                                       

```

```python
build_chart(gen_df).head()
```
