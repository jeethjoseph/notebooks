```python
#Packages
import pandas as pd
```

```python
#data exlporation
data = pd.read_csv('Data/1LakhFMCGSalesWithCategory.csv', names = ['BillId','ItemId','ItemName','Level1','Level2','Level3','Level4','Level5','Level6'] )
```

```python
data['dummy'] = 1
```

```python
data = data.dropna()
```

```python
matrix = data.pivot_table(values='dummy',index ='BillId', columns ='ItemName').fillna(0)
matrix.head()
```

```python
matrix.any().any() > 1
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(matrix, min_support=0.005, use_colnames=True)
```

```python
frequent_itemsets.tail()
```

```python
frequent_itemsets.shape
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.tail()
```
