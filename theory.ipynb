```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 3 visually seperable clusters
X,y = make_blobs(n_samples=300, centers=3, cluster_std=0.50, random_state=0)
```

```python
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
# Import the K-means Class
from sklearn.cluster import KMeans
```

```python
#Initialise the object, clusters =3 centeoid initialisation is random, max iter=10
kmeans = KMeans(n_clusters=3, init='random', max_iter=10)
```

```python
kmeans.fit(X)
```

```python
#predict class for every point
y_pred = kmeans.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred, s=50)
```

```python
centroids =kmeans.cluster_centers_
```

```python
#plot centroids
plt.scatter(X[:,0],X[:,1], c=y_pred, s=50)
plt.scatter(centroids[:,0],centroids[:,1],c='black' ,s=100, marker='X')
```

```python
#List that will hold sSS values for different cluster sizes
ss=[]

#computing ss for cluster sizes between 1 and 8
for i in range(1,9):
    kmeans = KMeans(n_clusters=i, random_state=0, max_iter=10, init='random').fit(X)
    #Append the values of ss for an iteration to sslist
    ss.append(kmeans.inertia_)
    
#Plot the elbow
sns.pointplot(x=[j for j in range(1,9)], y=ss)
    
```

# Half Moon clusters

```python
#import
from sklearn.datasets import make_moons

#Get access to points using ,make_moons

X_m, y_m = make_moons (200, noise=.05, random_state=0)

#Plot the tow halfs
plt.scatter(X_m[:,0],X_m[:,1], s=50)
```

# Apply k-means algorithm to make sense of clusters

```python
#Initialize K-Means with k=2
kmm =KMeans(n_clusters=2, init='random', max_iter=10)
kmm.fit(X_m)

#predicting classes
y_m_pred = kmm.predict(X_m)

plt.scatter(X_m[:,0],X_m[:,1], c=y_m_pred, s=50)
```

# Spectral clustering

```python
#Import Spectral Clustering
from sklearn.cluster import SpectralClustering

#Define the model
model = SpectralClustering(n_clusters = 2, affinity='nearest_neighbors')

#Fit and predict
y_m_sc = model.fit_predict(X_m)

plt.scatter(X_m[:,0], X_m[:,1], c=y_m_sc, s=50)
```
