# Prep done use SVD

```python
from surprise import SVD, Reader, Dataset, evaluate

reader = Reader()
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader) #need evaluation
data.split(n_folds=5)
svd = SVD()
trainset =data.build_full_trainset()
svd.train(trainset)
```

```python
svd.predict(2, 522)
```

```python
evaluate(svd, data, measures =['RMSE'])
```

```python
# Final Function
# Takes title, userId and returns to 25 sorted movie names on est
def reco(userId,title):
    idx =cosine_sim_map.loc[title,'movieId']
    df = cosine_sim.loc[idx].to_frame(name=None)
    df = df.sort_values(by =[idx],ascending = False).head(26)
    df2 = df.merge(movienamepicker, how ='left', on = 'movieId' ) 
    df3 = df2.drop(['id'], axis=1)
    df4 = df3.reset_index()
    df4['est'] = df4['movieId'].apply(lambda x: svd.predict(userId, x).est)
    df4 =df4.sort_values('est', ascending=False)
    
    return df4.head(10)

```
