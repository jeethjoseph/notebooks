```python
#Packages
import pandas as pd
```

```python
#data exlporation
data = pd.read_csv('Data/TenLakhLines_24Apr.csv', names = ['BillId','ItemId','ItemName','Level1','Level2','Level3','Level4','Level5','Level6'] )
```

```python
data['dummy'] = 1
```

```python
data['Level5'].unique()
```

```python
data = data.dropna()
```

```python
matrix = data.pivot_table(values='dummy',index ='BillId', columns ='Level5').fillna(0)
matrix.head()
```

```python
matrix.any().any() > 1
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(matrix, min_support=0.0005, use_colnames=True)
```

```python
frequent_itemsets.tail()
```

```python
frequent_itemsets.shape
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules.to_csv('Data/out.csv')
```

```python
toothrules = rules[rules["antecedents"] == frozenset(['SANITARY NAPKINS','CONTRACEPTIVE'])]
```

```python
toothrules = rules[rules["antecedents"] == frozenset(['CONTRACEPTIVE'])]
```
