```python
data['Description'] = data['Description'].str.strip()
```

```python
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
```

```python
data['InvoiceNo'].sort_values(ascending = False)
```

```python
#Credit Bill removal
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
#Still retains Invoices with A
#data = data[~data['InvoiceNo'].str.contains('A')]
```

```python
data['Country'].unique()
```

```python
#Taking only france data
#basket = data(data['Country'] == 'France')
basket = data[data['Country'] == 'France']
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo'])
```

```python
# Group by InvoiceNo and Description
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity']
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack()
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index()
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0)
```

```python
basket = data[data['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
```

```python
basket.head()
```

```python
#dummify
#All positive values are made 1. All negative values are made 0.
```

```python
def encode(x):
    if x <= 0:
        return 0
    else :
        return 1
```

```python
basketSets = basket.applymap(encode)
```

```python
basketSets.head()
```

```python
basketSets.any().any() > .2
```

```python
#Dont know what happens
basketSets.drop('POSTAGE', inplace=True, axis=1)
```

```python
basketSets.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(basketSets, min_support=0.07, use_colnames=True)
```

```python
frequent_itemsets.tail()
```

```python
frequent_itemsets.loc[50]['itemsets']
```

```python
frequent_itemsets.shape
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.tail()
```
